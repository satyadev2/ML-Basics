1. Initialize the Data


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x=2*np.random.rand(100,1)
y=4+3*x+np.random.randn(100,1)
```

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Generated Data')
plt.show()
```

2. Initialize Parameters

```python
w=0
b=0
learning_rate=0.01
epochs=200
```

3. Define the Hypothesis

```python
def predict(x,w,b):
    return np.dot(w,x)+b
```

4. Define the Cost Function

```python
def MSE(x,y,w,b):
    m=len(x)
    totalcost=(1/(2*m))*np.sum((predict(x,w,b)-y)**2)
    return totalcost
```

5. Perform Gradient Descent

```python
def gradient_descent(x,y,w,b,learning_rate,epochs):
    m=len(x)
    
    history = []

    for _ in range(epochs):
        predictions=predict(x,w,b)
        history.append(MSE(x, y, w, b))
        b-=learning_rate*(1/m)*np.sum(predictions-y)
        w-=learning_rate*(1/m)*np.sum((predictions-y)*x)

    print(f'Loss = {MSE(x, y, w, b)}')

    plt.plot(history)
    plt.show()
    return w,b
```

7. Visualize the Result

```python
w,b=gradient_descent(x,y,w,b,learning_rate,epochs)
```

```python
print(f"Trained parameters: w = {w}, b = {b}")
```

```python
# Plot the data and the regression line
plt.scatter(x, y, color='blue')
plt.plot(x, predict(x, w, b), color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.show()
```
