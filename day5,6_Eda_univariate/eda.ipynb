```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic=pd.read_csv('train.csv')
print(titanic.sample(2))
```

# 1.Categorical data

## countplot

```python
# sns.countplot(titanic,x='Survived')
titanic['Survived'].value_counts().plot(kind='bar')
# sns.countplot(titanic,x='Pclass')
```

## pie Chart

```python
titanic['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

# 2.Numerical Data

## Histogram

```python
# ''' bins are created '''
plt.hist(titanic['Age'],bins=15)
```

## Distplot -- skewness

```python
#improvement over histogram --
sns.displot(titanic['Age'],kde=1)
```

## Boxplot ---used for 5 number summary

```python
sns.boxplot(titanic['Fare'])
```

```python
#min, max, mean, median, variance
print(titanic['Age'].min())
```
